```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print('modules are imported')
```

```python
corona_dataset=pd.read_csv("covid19_Confirmed_dataset.csv")
corona_dataset.head(10)
```

```python
corona_dataset.shape
```

```python
df= corona_dataset.drop(["Lat","Long"],axis=1,inplace=True)
corona_dataset.head(10)
```

```python
data_aggregated=corona_dataset.groupby("Country/Region").sum()
data_aggregated.head()
```

```python
data_aggregated.shape
```

```python
data_aggregated.loc["India"].plot()
data_aggregated.loc["Italy"].plot()
data_aggregated.loc["Spain"].plot()
plt.legend()
```

```python
data_aggregated.loc["India"][43:46].plot()
plt.legend()
```

```python
data_aggregated.loc["India"].diff().plot()
```

```python
data_aggregated.loc["India"].diff().max()
```

```python
data_aggregated.loc["Italy"].diff().max()
```

```python
countries= list(data_aggregated.index)
max_infection_rates=[]
for c in countries :
    max_infection_rates.append(data_aggregated.loc[c].diff().max())
data_aggregated["max_infection_rates"]=max_infection_rates
```

```python
data_aggregated
```

```python
corona_data=pd.DataFrame(data_aggregated["max_infection_rates"])
```

```python
corona_data
```
